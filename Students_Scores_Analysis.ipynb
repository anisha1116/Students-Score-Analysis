# Data Analysis Project on Student's Score Analysis

```python
import numpy as np
import pandas as pd
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("students_score.csv")
print(df.head())
```

```python
df.describe()
```

```python
df.info()
```

```python
df.isnull().sum()
```

# Drop unnamed column

```python
df = df.drop("Unnamed: 0", axis = 1)
```

```python
df.head()
```

# Gender Distribution 

```python
plt.figure(figsize=(4,5))
ab = sns.countplot(data = df, x = "Gender")
ab.bar_label(ab.containers[0])
plt.title("Gender Distribution")
plt.show()
```

```python
# conclusion1: number of females is more than the number of males
```

```python
gb1 = df.groupby("ParentEduc").agg({"MathScore":'mean',"ReadingScore":'mean',"WritingScore":"mean"})
print(gb1)
```

```python
sns.heatmap(gb1, annot = True)
plt.title("Relationship betweeen Parent's education and Student's scores")
plt.show()
```

```python
# conclusion2: education of the parents have a good impact on the student's scores
```

```python
gb2 = df.groupby("ParentMaritalStatus").agg({"MathScore":'mean',"ReadingScore":'mean',"WritingScore":"mean"})
print(gb2)
```

```python
plt.figure(figsize=(4,4))
sns.heatmap(gb2, annot = True)
plt.title("Relationship betweeen Parent's Marital status and Student's scores")
plt.show()
```

```python
# conclusion3: marital status of the parents have negligible impact on student's scores
```

# Distribution of Ethnic Group

```python
print(df["EthnicGroup"].unique())
```

```python
groupA = df.loc[(df["EthnicGroup"] == "group A")].count()
print(groupA)
```

```python
groupA = df.loc[(df["EthnicGroup"] == "group A")].count()
groupB = df.loc[(df["EthnicGroup"] == "group B")].count()
groupC = df.loc[(df["EthnicGroup"] == "group C")].count()
groupD = df.loc[(df["EthnicGroup"] == "group D")].count()
groupE = df.loc[(df["EthnicGroup"] == "group E")].count()

mlist = [groupA["EthnicGroup"],groupB["EthnicGroup"],groupC["EthnicGroup"],groupD["EthnicGroup"],groupE["EthnicGroup"]]
print(mlist)
```

```python
lab = ["groupA", "groupB", "groupC", "groupD", "groupE"]
plt.pie(mlist , labels = lab, autopct="%.0f%%")
plt.title("Distribution of Groups")
plt.show()
```
